- Edge Detection 실험

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
import os
from glob import glob

IMAGE_DIR = os.path.join(os.getcwd(), 'datasets')
NEW_DATA = os.path.join(IMAGE_DIR, 'new_tire')
OLD_DATA = os.path.join(IMAGE_DIR, 'old_tire')

new_paths = glob(os.path.join(NEW_DATA, '*.jpg'), recursive=True)
old_paths = glob(os.path.join(OLD_DATA, '*.jpg'), recursive=True)

print(old_paths)

PLT_ROW = len(new_paths)
plt.figure(figsize=(15,20))

for idx, path in enumerate(new_paths):
# for idx, path in enumerate(old_paths):
    idx *= 2

    img = cv.imread(path,0)
    edges = cv.Canny(img,20,80)

    plt.subplot(PLT_ROW, 2, idx+1)
    plt.imshow(img,cmap = 'gray')
    plt.title('Original Image', fontsize=9)

    # plt.xticks([]), plt.yticks([])

    plt.subplot(PLT_ROW, 2, idx+2)
    plt.imshow(edges,cmap = 'gray')
    plt.title('Edge Image', fontsize=9)
    # plt.xticks([]), plt.yticks([])

plt.suptitle('Threshold: 20 ~ 80', fontweight='bold')
plt.savefig('세번째 자료.jpg')
```
